```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

result = []

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.figure(dpi=100)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '지역의 남녀 성별 인구 차이 분포')
plt.bar(range(101), result)
plt.show()

# 대전 50대 50으로 남녀의 비율이 같았음.
# 남녀 성별 인구 차이 분포도 이와같이 중앙을 기점으로 상하반전된 것같은 그래프를 보여줌
# 20대 초반만 남성 비율이 적음 -> 군대를 빨리가는것으로.. 예상
# 대한민국 교통의 중심지라 많은 유입과 전출이 있어 중장년층의 성비가 비슷할 것이라 예상
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

result = []

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.figure(dpi=100)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '지역의 남녀 성별 인구 차이 분포')
plt.bar(range(101), result)
plt.show()

# 울산은 자동차·조선 및 석유화학 공업이 발달.
# --> 공업이 발달된 도시이기에 60대까지 몇몇 구간을 제외하고 남성의 인구가 월등히 높을 것으로 예상

# 20대 초중반 남성 많은 이유 - >현대자동차나 공장 등 단기 아르바이트를 많이해서 그럴 수 있을 것 같음

# 20대 후반부터 남성의 비율이 점점 줄어드는데 뉴스에 따르면 청년층의 이탈이 많아짐. 
# 중화학공업이 중심인 도시라 블루컬러 대신 화이트컬러 일자리를 찾아 떠나는 남성 청년층이 많을 것으로 예상

# 1970~80년대 중화학공업이 막 발달할 무렵 근무했던 노동자들이 60대까지 근무할 것으로 예상.


```
