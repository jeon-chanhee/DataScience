```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('plotting')  # 타이틀 속성 추가

plt.plot([10,20,30,40], label='asc') # 증가를 의미하는 asc 범례
plt.plot([40,30,20,10], label='desc') # 감소를 의미하는 desc 범례

plt.legend(loc=5) # 그래프에 범례 속성 추가를 위해 꼭 넣어줘야함 
plt.show()
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('linestyle')  # 타이틀 속성 추가

plt.plot([10,20,30,40], color='r', linestyle = '--',label='dashed') # 빨산색 dashed 그래프
plt.plot([40,30,20,10], 'g', ls = ':',label='dotted') # 초록색 dotted 그래프 linestyle -> ls 표현 가능

plt.legend() # 그래프에 범례 속성 추가를 위해 꼭 넣어줘야함 
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)

result = [] # 결측치를 제거한 최고 기온 리스트

for row in data:
  if row[-1] != '':   # 최고 기온이 ''(빈문자열)이 아니면, result에 저장
    result.append(float(row[-1]))  # 실수로 변환

result

plt.figure(dpi=200)   # dpi는 해상도
plt.plot(result,'r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

max_rain = -999 
max_date = ''


f = open('/content/sejong_rainy.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)



for row in data:
  if row[-1] == '':
    row[-1] = -999
  row[-1] = float(row[-1])  
  
  if max_rain < row[-1]:
    max_date = row[0]
    max_rain = row[-1]

f.close()

print('세종시에서 강수량이 가장 많았던 날은' , max_date+'로',max_rain,'mm이였습니다.')
```
