```python
# 주사위 5번 던져서 나온 데이터 최종 결과로 한번에 출력

import random
import matplotlib.pyplot as plt

dice = [] # 주사위 리스트 변수 선언

# 왜 list에 담았나? 0부터 순서대로 데이터를 담고 한번에 출력하려고

for i in range(5):
  dice.append(random.randint(1,6))

# 랜덤하게 굴린 주사위 5개의 값을 히스토그램으로 표현
plt.figure(dpi=200)
plt.hist(dice, bins=6)        # bins 는 구간 정하기 6까지
plt.show()
```

1월과 8월의 데이터를 히스토그램으로 시각화 하

```python
import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더

aug = [] # 8월 최고 기온 담을 리스트 변수 선언
jan = [] # 1월의 최고 기온을 담을 리스트 변수 선언
june = [] # 6월

for row in data:
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터 추출
  if row[-1] != '':
    if month == '08':
      aug.append(float(row[-1]))
    if month == '01':
      jan.append(float(row[-1]))
    if month == '06':
      june.append(float(row[-1]))

# 최고 기온 데이터의 히스토그램
plt.figure(dpi=200)
plt.hist(aug, bins=10, color = 'r',label='Aug')
plt.hist(jan,bins=10, color='b',label='Jan')
plt.hist(june,bins=10, color='g',label='June')
plt.legend() 
plt.show()
```

1월부터 12월까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더

month = [[],[],[],[],[],[],[],[],[],[],[],[]]  # 월별 데이터 저장할 month 변수 선언

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1])) # int 형으로 만들어서 숫자로 만듬 그 다음에 -1 ex) 1월이면 0번째 방 2월이면 1번째 방...

plt.figure(dpi=200)
plt.boxplot(month)
plt.show()
```

8월 1일부터 31일까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) 

day = [] # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31):
  day.append([])   # day 리스트 내에 31개 리스트 생성 [[],[],[]...]

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1])) # 8월 최고 기온 값 저장 day[0] -> 1일

plt.figure(figsize=(10,5), dpi=200)
plt.boxplot(day,showfliers=False) # 아웃라이어(이상치) 값 생략
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) 

day = [] # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31):
  day.append([])   # day 리스트 내에 31개 리스트 생성 [[],[],[]...]

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1])) # 8월 최고 기온 값 저장 day[0] -> 1일

plt.style.use('ggplot') # 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)
plt.figure(figsize=(10,5), dpi=200)
plt.boxplot(day,showfliers=False) # 아웃라이어(이상치) 값 생략
plt.show()
```
