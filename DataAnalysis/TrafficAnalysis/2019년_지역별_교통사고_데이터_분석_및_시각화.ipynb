```python
import plotly.express as px

import pandas as pd
```

데이터 준비하기

```python
df = pd.read_csv('/content/도로교통공단_교통사고 정보.csv', encoding='euc-kr')
df.head()
```

```python
df.info()
```

교통사고 시간 전처리

```python
#Substr 문자열 자르기 위해 데이터타입 변경
df = df.astype({'발생년월일시':'string'})

# --> split을 사용할 수 도 있기 때문에 (일시만 살펴보고자 할때)
```

```python
df.info()
```

```python
#2019010100 -> 00 시간으로 분리
df['발생시간'] = df['발생년월일시'].str[8:]   # 문자열 슬라이싱을 통한 열 추가
```

```python
df.info()
```

```python
 #발생시간 숫자 데이터타입으로 변경
 df = df.astype({'발생시간':'int64'})   # 숫자로 바꿔주기
 # 시간으로 차트를 그릴 수 있음
```

```python
# 2019010100 -> 20190101 로 추출 후  -> 2019-01-01 (날짜형태 YYYY-MM-DD 형태로 변경)
df['발생년월일시'] = pd.to_datetime(df['발생년월일시'].str[:8],
                              format='%Y-%m-%d' , errors='raise')  # 발생년월까지만 문자열 뽑아내라 
```

```python
df.info()
```

```python
fig = px.scatter(df,x='발생년월일시',y='발생시간',color='발생지시도',size='사망자수',hover_data=['발생시간'])
fig.show()
```

지도 활용하여 교통사고 현황 분석

```python
import folium

map = folium.Map(location=[36.351,127.386])  # 각 도시 시청을 기준으로함
map
```

지도에 필요한 정보만 담기 -> 대전의 교통사고 현황을 지도에 표출

```python
df_map = df[df.발생지시도=='대전']
df_map.info()
```

사망자수, 부상자수, 중상자수, 경상자수 전처리 -> 지도에 원을 그리기 위해서

```python
df_map = df_map.astype({'사망자수':'float64'})
df_map = df_map.astype({'경상자수':'float64'})
df_map = df_map.astype({'중상자수':'float64'})
df_map = df_map.astype({'부상자수':'float64'})
```

```python
df_map.info()
```

사망자수, 경상자 수에 따른 CircleMarker 지도에 표출하기

```python
for n in df_map.index:
  cnt = df_map['사망자수'][n] + df_map['경상자수'][n] # 사망자수 + 경상자수 -> CircleMarker 사이즈
  folium.CircleMarker([df_map['위도'][n], df_map['경도'][n]],
                       radius=cnt*10, popup=df_map['사고유형'][n], color='#3186cc',fill_color='#3186cc').add_to(map)

map.save('대전교통사고현황.html')
```
