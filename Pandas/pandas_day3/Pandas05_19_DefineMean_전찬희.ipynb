```python
datalist = [28,31,24,25,30,32,20,30,31,26,31]
print(datalist)

def mMean():
    add =0
    for i in datalist:
        add = add+i
        mean = add/len(datalist)
    return mean                    # return 값을 쓰면 함수를 계속 재사용할 수 있음 값을 출력하기 때문에

print('평균 : %d' % mMean())

# Data Sort(오름차순으로) 홀수개이면 가운데값
# Data Sort 짝수개이면 2개의 평균값 구하기

def mMedian():
    datalist.sort()
    odd = int((len(datalist)+1)/2)    # 2로 나누면 실수가 되기 때문에 int로 정수형으로 바꿔줘야함
    jjak = int(len(datalist)/2)
    midodd = datalist[odd]
    midjjak = (datalist[jjak]+datalist[odd])/2
    if len(datalist) % 2 == 1:
        return midodd
    else:
        return midjjak

print('중앙값: ' , mMedian())

# 관측값 - 관측값 평균
def mDeviation():
    devList = []
    datalist.sort()
    avgMean = mMean()       # 함수를 그대로 쓰기보다 변수에 넣어서 쓰는게 안전해서
    for i in datalist:
        devList.append(i-avgMean)
    return devList
    
        

print('편차 :', mDeviation())          # 리스트 출력값을 보낼라고


# 분산은 편차의 제곱/관측수 -1
def mVariance():
    varList = []
    datalist.sort()
    dev = mDeviation()
    for i in dev:
        varList.append(i**2)
    square_var = sum(varList)/(len(dev)-1)
    return square_var

print('분산 : %0.1f' % mVariance())
        
#  문제4] 표준편차 : 분산의 제곱근
def mStandardDeviaton():
    mSD = mVariance()
    return mSD**(1/2)
    

print('표준편차 : %0.2f' % mStandardDeviaton())



# 문제5] 범위(range) : 최대값과 최소값의 차이
import pandas
def mRange():
    a = pandas.Series(datalist)
    #a = pandas.Series([28,31,24,25,30,32,20,30,31,26,31])
    mMax = a.max()
    mMin = a.min()
    return mMax-mMin

print('범위 :', mRange())

#  devList2 = [i-avgMean for i in datalist]
```
