```python
import pandas as pd

df=pd.read_csv('./../data/gapminder.tsv',sep='\t')
```

```python
# 2중 포문 사용해서 2중 groupby 만들기
uniList = df['year'].unique()
for idx in uniList:
    print('%s년'% idx,'\n')
    grYear = df[df['year']==idx]
    conList = grYear['continent'].unique()
    for idx2 in conList:
        grCon = grYear[df['continent']==idx2]  # continent를 grYear에서 찾아보겟다
        print('지역 : %s   평균 : %f' % (idx2,grCon['lifeExp'].mean()))
        
       
    print()
    
```
