```python
import pandas as pd

df=pd.read_csv('./../data/gapminder.tsv',sep='\t')
```

```python
uniList = df['year'].unique()
for idx in uniList:
    print('년도(인덱스) : ',idx,'\n=====> 1')
    grYear = df[df['year'] == idx]
    print('groupby된 연도의 개수 : ',len(grYear),'\n======> 2',grYear.head(),'\n====> 3','\n행,열의 갯수',grYear.shape)
    print('평균:', grYear['pop'].mean())
    print()
```
