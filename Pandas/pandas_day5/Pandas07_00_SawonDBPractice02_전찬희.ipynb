```python
import pandas as pd

deptDB = pd.read_csv('./../data/deptDB.csv',header=None)
deptDB.columns=['deptno','dname','loc']

deptDB = deptDB.replace('\'','',regex=True)
print(deptDB)
```

```python
sawonDB = pd.read_csv('./../data/sawonDB.csv',header=None)
sawonDB.columns=['sabun','saname','deptno','sajob','sapay','sahire','sasex','samgr']
sawonDB = sawonDB.replace('\'','',regex=True)
print(sawonDB)
```

```python
# 문제01] SawonDB 데이터에서 입사년도가 88년도인 사람 출력
for i in range(len(sawonDB)):
    if sawonDB.iloc[:,5][i][1:5] =='1988':  # 1988 이전에 ' 이라는 문자가 있기에 1:5를 범위로 지정
        print(sawonDB.iloc[i,:8])
```

```python
# 문제04] SawonDB 데이터에서 직위별 급여 평균 출력
print(sawonDB.groupby('sajob')['sapay'].mean())
```

```python
# 문제 06] 인상급여:sawonDB 데이터에서 급여 10% 인상, 인상급여 field 생성
# 문제 07] sawonDB 데이터에서 전산부 직원의 평균연봉 출력
# 문제 08] 컬럼 이름순 재배치, sort_values() 함수 적용
# 문제 09] sawonDB 데이터에서 직위가 사원과 대리중 직원수가 4인이상인 직위, 평균급여 출력
```

```python
# 문제 06] 인상급여:sawonDB 데이터에서 급여 10% 인상, 인상급여 field 생성
data = sawonDB.loc[:,'sapay']*1.1
sawonDB['sapayUp'] = data
sawonDB
```

```python
# 문제 07] sawonDB 데이터에서 전산부 직원의 평균연봉 출력
sawonDB.loc[sawonDB['deptno']==30]['sapay'].mean()
print(sawonDB.loc[sawonDB['deptno']==30]['sapay'])
print('전산부 직원 평균연봉 : ',sawonDB.loc[sawonDB['deptno']==30]['sapay'].mean())
```

```python
# 문제 08] 컬럼 이름순 재배치, sort_values() 함수 적용
sawonDB.sort_values(['saname'])
```
