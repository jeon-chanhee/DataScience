출근 시간 (7시-9시) 막대 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')

data = csv.reader(f)
next(data)
next(data)

result = [] # 출근시간 7시-9시 승차 데이터 추출 -> 10,12,14번째 인덱스에 존재

for row in data:
  row[4:] = map(int, row[4:])  # 따옴표로 묶여있던 숫자들 정수형으로 바뀜
  result.append(sum(row[10:15:2]))   # 10,12,14 추출 후 합침

result.sort() # 데이터를 오름차순으로 정렬

# 데이터 시각화
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.bar(range(len(result)),result)
plt.show()
```

밤 11시에 사람들이 가장 많이 타는 역은 어디일까


승차 패턴 찾기<br> 
승차 시각이 1시간씩 늦어질 때 마다 인덱스 2씩 늘어남 <br> 승차 시작 시간의 인덱스 값 4 -> 4시부터 승차가 시작(첫차)<br> 
승차 시각(t) 와 승차 인원이 적힌 인덱스(i) 관계 분석 <br>
i = 4 + (t -4) * 2


```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')

data = csv.reader(f)
next(data)
next(data)

mx = 0
mx_station = ''

t = int(input('몇 시에 타는지 궁금하세요? :'))

for row in data:
  row[4:] = map(int, row[4:])
  a = row[4 + (t-4)*2] # 입력 받은 시각의 승차 인원 값 추출
  if a > mx:
    mx = a
    mx_station = row[3] + '(' + row[1] + ')'

print(mx_station,mx)
```

시간대별로 하차 인웡니 가장 많은 역 막대 그래프로 그리

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')

data = csv.reader(f)
next(data)
next(data)

mx = [0] * 24
mx_station = [''] * 24

for row in data:
  row[4:] = map(int, row[4:])
  for j in range(24):
    a = row[5 + j * 2]  # 0부터 돌기때문에 새벽 4시 인덱스부터 시작, 시간대별 패턴 i = j * 2 + 4 (j는 0 ~ 23까지)
    if a > mx[j]:
      mx[j] = a
      mx_station[j] = row[3] + '(' + str(j+4)+')'   # 역 이름과 시간대 추가


plt.rc('font', family='NanumBarunGothic')
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.bar(range(24), mx)
plt.xticks(range(24),mx_station, rotation=90)
plt.show()
```

모든 지하철역에서 시간대별 승하차 인원 모두 더하면

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')

data = csv.reader(f)
next(data)
next(data)

s_in = [0] * 24
s_out = [0] * 24

for row in data:
  row[4:] = map(int, row[4:])
  for i in range(24):
    s_in[i] += row[4 + i * 2] # 승차
    s_out[i] += row[5 + i * 2] # 하차

plt.rc('font', family='NanumBarunGothic')
plt.figure(dpi=200)
plt.title('지하철 시간대별 승하차 인원 추이')
plt.style.use('ggplot')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24),range(4,28))
plt.show()
```
