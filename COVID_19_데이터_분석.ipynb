데이터 준비하기

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df_patient = pd.read_csv('PatientInfo.csv')
```

```python
df_patient.head()
```

```python
df_patient.tail()
```

```python
df_patient.info()
```

데이터 결측치 확인

```python
df_patient.isna().sum()     # isna() 결측치 true,false 구분 , sum() false인 값들 더하기
```

나이 (age) -> 20s -> 23 정확하게 숫자로 변경 -> 현재년도 - 태어난 년도(birthday_year)<br>
1. Null (Nan) 데이터를 처리 (계산 불가) -> 0으로 채우자 <br>
2. 나이 계산 시, 0보다 큰 값은 현재년도 - 태어난 년도 0보다 작으면 다시 Nan으로 채우기

```python
df_patient['birth_year'] = df_patient.birth_year.fillna(0.0).astype(int)  # birth_year nan값 0으로 채우기
df_patient.tail()
```

```python
# 0보다 큰 값은 그대로 사용하고, 0인 값은 Nan으로 변환 -> 나이 계산 시 2021살
df_patient['birth_year'] = df_patient['birth_year'].map(lambda val : val if val > 0 else np.nan)
# np.nan으로 nan의 타입을 바꿔주는 것.. 데이터타입을 명확히 하기 위해
# lambda 결과물을 리스트로 만들어주는것
# lambda 인자 : 표현식
# map은 리스트의 요소를 지정된 함수로 처리해주는 함수
df_patient.info() 
```

```python
df_patient['age'] = 2021 - df_patient['birth_year']
df_patient.tail()
# NaN 값이 나오는건 df_patient['birth_year'] 0의 값이 Nan 값이 되었기 때문에 계산이 불가능해서
# 이 과정을 안하면 2021-0 = 2021살이 나오기 때문에 
# 결측치인 nan값을 버리기 위해서
```

```python
patient = df_patient  # 전처리한 원본을 살리기 위해서 patient라는 변수에 따로 담음
```

2. 격리해제자(released) 데이터

```python
# 격리해제자 정보 담기
released = patient[patient.state == 'released']
released.head()
```

```python
released.info()
```

격리해제자들의 연령대 분석

```python
# 격리해제자들의 연령대 분포
plt.figure(figsize=(10,6),dpi=200)
sns.set_style('darkgrid')
plt.title('Age distribution of the released')
sns.kdeplot(data=released['age'],shade=True)
```

3. 격리된 환자 데이터

```python
# 격리된 환자 정보 담기
isolated_state = patient[patient.state == 'isolated']
isolated_state.head()
```

```python
isolated_state.info()
```

```python
# 격리된 환자들의 연령대 분포
plt.figure(figsize=(10,6),dpi=200)
sns.set_style('darkgrid')
plt.title('Age distribution of the released')
sns.kdeplot(data=isolated_state['age'],shade=True)
```

4. 사망자 데이터 

```python
# 사망자의 정보
dead = patient[patient.state=='deceased']
dead.info()
```

```python
# 사망한 환자들의 연령대 분포
plt.figure(figsize=(10,6),dpi=200)
sns.set_style('darkgrid')
plt.title('Age distribution of the released')
sns.kdeplot(data=dead['age'],shade=True)
```

5. 사망자에 따른 성별 데이터

```python
male_dead = dead[dead.gender=='male']
female_dead = dead[dead.gender=='female']
```

```python
male_dead.head()
```

```python
female_dead.info()
```

성별에 따른 사망자의 연령분포

```python
# 성별에 따른 사망자의 연령대 분포
plt.figure(figsize=(10,6),dpi=200)
sns.set_style('darkgrid')
plt.title('Age distribution of the deceased the gender')
sns.kdeplot(data=male_dead['age'],label='Male',shade=True)
sns.kdeplot(data=female_dead['age'],label = 'Female',shade=True)
plt.legend()
```

KDE : Kernel Density Estimator 추가한 distplot()

```python
# 좀 더 명확한 추이를 볼 수 있는 그래프
plt.figure(figsize=(10,6),dpi=200)
sns.set_style('darkgrid')
plt.title('Age distribution of the deceased the gender')
sns.distplot(a=male_dead['age'],label='Male',kde=True)
sns.distplot(a=female_dead['age'],label = 'Female',kde=True)
plt.legend()
```

사망자 성별 비교 -> 바차트, 파이차트 -> 분포

```python
plt.figure(figsize=(10,6),dpi=200)
plt.title('Gender')
dead.gender.value_counts().plot.bar()   # gender 안의 수를 세줌
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.title('Gender')
dead.gender.value_counts().plot.pie(autopct = '%.1f%%')  # 비율 확인하기 
```

환자 유형별 (격리해제자, 격리중 환자, 사망자) 연령 분포 

```python
# 성별에 따른 사망자의 연령대 분포
plt.figure(figsize=(10,6),dpi=200)
sns.set_style('darkgrid')
plt.title('Age distribution of the deceased the gender')
sns.kdeplot(data=dead['age'],label='deceased',shade=True)
sns.kdeplot(data=released['age'],label = 'released',shade=True)
sns.kdeplot(data=isolated_state['age'],label = 'isolated',shade=True)
plt.legend()
```

환자 유형별 (격리해제자, 격리중 환자, 사망자) 연령 분포 파이 차트로 그리기

```python
# 성별에 따른 사망자의 연령대 분포

plt.figure(figsize=(10,6),dpi=200)
plt.title("Distribution of the patient's state ")
patient.state.value_counts().plot.pie(autopct = '%.1f%%')
```

```python
# 성별,날짜별 사망자 수 꺾은선 그래프
plt.figure(figsize=(10,6),dpi=200)
sns.set_style('darkgrid')
plt.title('Age distribution of the deceased the gender')
# 날짜는 무조건 datetime이라는 날짜로 취급해야함
pd.to_datetime(male_dead.confirmed_date).value_counts().plot(label='male_dead')
pd.to_datetime(female_dead.confirmed_date).value_counts().plot(label='female_dead')
plt.legend()
```
