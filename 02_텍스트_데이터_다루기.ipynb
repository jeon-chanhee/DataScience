```python
import csv                      # 경로 복사해서 파일명 삽입하기
with open("/content/Demographic_Statistics_By_Zip_Code.csv") as infile:
  data = list(csv.reader(infile))
```

```python
# 첫번째 레코드에 COUNT PARTICIPANTS 데이터 인덱스값 추출 ---> data[0] 컬럼 값
countParticiapntsindex = data[0].index("COUNT PARTICIPANTS")     # 데이터 0번째 즉, 컬럼들중 "COUNT PARTICIPANTS"의 인덱스를 알려줘
countParticiapntsindex
# 데이터 정제할때 데이터 바뀔 수 있기 때문에 인덱스 값 확인
```

나이의 평균과 표준편차

```python
import statistics

print(statistics.mean(countParticiapnts),statistics.stdev(countParticiapnts))  # 평균, 표준편차
# pandas를 더 많이 사용
```

```python
wn = nltk.corpus.wordnet # 코퍼스 리더(reader)

# synset :유의어 묶음으로 단어-품사-순번 으로 구성
wn.synsets("cat")
```

```python
wn.synset('cat.n.01').hypernyms()   # hypernyms 상위어
wn.synset('cat.n.01').hyponyms()   # hyponyms 하위어
```

Wordnet 사용해서 synset간 의미론적 유사도 계산하기 (0-1 사이 실수)

0이면 두 단어 서로 관계 없음
1이면 완전한 유의어



```python
# 고양이, 링스 : 시라소니 유사도 계산
x = wn.synset('cat.n.01') # 고양이
y = wn.synset('lynx.n.01') # 시라소니

# x,y 유사도 계산
x.path_similarity(y)
```

```python
x = wn.synset('cat.n.01') # 고양이
y = wn.synset('domestic_cat.n.01') # 집고양이
x.path_similarity(y)
```

```python
x = wn.synset('cat.n.01') # 고양이
y = wn.synset('dog.n.01') # 집고양이
x.path_similarity(y)
```

```python
# 1. 토큰화 (텍스트 -> 단어로 쪼갠다.)

from nltk.tokenize import WordPunctTokenizer
word_punct = WordPunctTokenizer()

text = "}Help! :))) :[ ..... :D{"

# Tokenizer는 모든 구두점 (punctuation; 문장부호)을 기준으로 분리
# -> 이모티콘을 이용한 감성 분석 등 문장구조 깊이 분석할 때 사용

word_punct.tokenize(text)
```

```python
# 2. 단어의 대소문자를 통일한다 (전부 다 대문자 or 소문자)
# 3. 불용어 제거 (stopwords 리스트 참조 THE 같은 것들 제거)
# 4. 형태소 분석(stemming) 단어를 형태소로 변환

# 포터 형태소 분석기 (보수적)

pstemmer = nltk.PorterStemmer()
pstemmer.stem("wonderful")
```

index.html 파일에서 (불용어 제외) 가장 많이 등장한 단어 원형 찾아보기

```python
# www.networksciencelab.com 사이트에서 하이퍼 링크 정보 모두 가져오기
from bs4 import BeautifulSoup
from urllib.request import urlopen
with urlopen("http://www.networksciencelab.com") as doc: # doc 라는 곳에 url 담음
  soup = BeautifulSoup(doc)

# 하이퍼 링크 추출
links = [(link.string, link["href"]) for link in soup.find_all("a") if link.has_attr("href")]



# 책 제목만 추출
html_text = [i[0] for i in links]    # i[0] 책 제목만 추출

# None 데이터 제거
html_text= " ".join(html_text[2:])   # none이 0,1 에 있으니 2부터 추출하자 -> 그러고 리스트에서 텍스트로 변환
                                     # . 없애고 공백 추가하니 텍스트가 됨
html_text 
```

```python
# index.html 파일로 저장
html_file = open("index.html",'w')    # w -> write 쓸려고 여는것
html_file.write(html_text)
html_file.close()
```

Index.html 파일에서(불용어 제외) 가장 많이 등장한 단어 원형 찾아보

```python
from bs4 import BeautifulSoup      # html 추출할때
from collections import Counter    # 단어를 추출해서 개수를 셀때
from nltk.corpus import stopwords  # 사전에서 불용어 제외
from nltk import LancasterStemmer  # 원형데이터 찾아내기

# 형태소 분류기를 생성한다.
ls = nltk.LancasterStemmer()

# 파일을 읽고 soup를 만든다.
with open("/content/index.html") as infile:
  soup = BeautifulSoup(infile)
```

```python
soup
```

```python
soup.text    # html 파일을 text 파일로 읽을 수 있게 하는것
```

```python
# 1) 텍스트 추출 -> 토큰화
words = nltk.word_tokenize(soup.text)     # 단어별로 자름.
words
```

```python
# 2) 단어를 소문자로 변환
words = [w.lower() for w in words]      # w 는 word 단어 1개씩 가져올것이기 때문에
                                        # 단어 소문자로 변환 w.lower()
words
```

```python
import nltk
nltk.download('stopwords')
```

```python
# 3) 불용어+ 특수기호를 제거하고 단어의 형태소를 추출
# 3-1) 불용어 제거 : stopwords.words("english")
# 3-2) 특수기호 제거 : isalnum()
words = [ls.stem(w) for w in words if w not in stopwords.words("english") and w.isalnum()] 
    #형태소로 변환                # 영어에 있는 불용어 제거        #isalnum() 문자열이 알파벳([a-zA-Z])과 숫자([0-9])로만 구성
words
```

```python
# 4) 가장 빈번하게 등장하는 단어 10개 추출
freqs = Counter(words)
print(freqs.most_common(10))
```
