데이터 준비하기

```python
# 최고 기온 데이터만 리스트에 담아서 출력하기

import csv 
f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더

result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '':
    result.append(float(row[-1]))

print(result)



```

```python
# 최고기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화

import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더

result = [] # 최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08': # 날짜를 잘라서 8월 데이터만 추출
      result.append(float(row[-1]))    # 결과 리스트에 담기

plt.figure(dpi=200)   # figsize=(가로10인치,세로10인치)
plt.plot(result, 'hotpink')
plt.show()
```

매년 생일 기준 최고 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더

result = [] 

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '24': # 매년 6월 9일
      result.append(float(row[-1]))    # 결과 리스트에 담기

plt.figure(dpi=200)   # figsize=(가로10인치,세로10인치)
plt.plot(result, 'hotpink')
plt.show()
```

매년 생일 기준 최고 기온, 최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더

high = [] # 최고 기온
low = [] # 최저 기온

for row in data:
  if row[-1] != '' and row[-2] != '':
    if row[0].split('-')[1] =='10' and row[0].split('-')[2]=='04':
      high.append(float(row[-1]))  # 최고 기온
      low.append(float(row[-2]))  # 최저 기온

plt.figure(dpi=200)   # figsize=(가로10인치,세로10인치)
plt.plot(high, 'r', label='High Tem') # 최고기온 
plt.plot(low, 'blue',label='Low Tem') # 최저 기온
plt.legend() # 범례 표시
plt.show()
```

날짜 구간을 좁혀서 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더

high = [] # 최고 기온
low = [] # 최저 기온

for row in data:
  if row[-1] != '' and row[-2] != '':
    if 1983 <= int(row[0].split('-')[0]):   # 숫자 비교를 위해 int로 감싼다
      if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '09':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
  
plt.figure(dpi=200)   # figsize=(가로10인치,세로10인치)
plt.plot(high, 'r', label='High Tem') # 최고기온 
plt.plot(low, 'blue',label='Low Tem') # 최저 기온
plt.legend() # 범례 표시
plt.show()
```

내 생일 주기

```python
import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더

high = [] # 최고 기온
low = [] # 최저 기온

for row in data:
  if row[-1] != '' and row[-2] != '':
    if 1995 <= int(row[0].split('-')[0]):   # 숫자 비교를 위해 int로 감싼다
      if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '12':
        high.append(float(row[-1]))
        low.append(float(row[-2]))

# 차트 설정  
plt.figure(dpi=200)   # 해상도
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# plot 차트 그리기
plt.plot(high, 'r', label='최고 기온') # 최고기온 
plt.plot(low, 'blue',label='최저 기온') # 최저 기온


plt.title('내 생일의 기온변화 그래프')
plt.legend() # 범례 표시

plt.show()
```

매년 생일 최고 기온, 평균 기온, 최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt 

f = open('/content/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더

high = [] # 최고 기온
low = [] # 최저 기온
avg = [] # 평균 기온

for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3] !='': # 최고 최저 평균 결측치 제거
    if 1995 <= int(row[0].split('-')[0]):   # 숫자 비교를 위해 int로 감싼다
      if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '12':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        avg.append(float(row[-3]))

# 차트 설정  
plt.figure(dpi=200)   # 해상도
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# plot 차트 그리기
plt.plot(high, 'r', label='최고 기온') # 최고기온 
plt.plot(low, 'blue',label='최저 기온') # 최저 기온
plt.plot(avg,'green',label ='평균 기온')


plt.title('내 생일의 기온변화 그래프')
plt.legend() # 범례 표시

plt.show()
```
