```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

male = []
female = []

for row in data:
  if '산남동' in row[0]:
    for i in range(0,101):
      male.append(int(row[i+3]))
      female.append(-int(row[-(i+1)]))

female.reverse()

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.title('산남동 지역의 남녀 성별 인구 분포')
plt.barh(range(101), male, color = 'skyblue',label = '남성')
plt.barh(range(101), female, color = 'hotpink', label = '여성')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt



f_2012 = open('2012.csv',encoding='cp949')
data_2012 = csv.reader(f_2012)
f_2021 = open('2021.csv',encoding='cp949')
data_2021 = csv.reader(f_2021)

result_2012 = []
result_2021 = []

for row in data_2012:
  if '한솔동' in row[0]:
    for i in row[3:]:
      result_2012.append(int(i.replace(',','')))  # -> 1000단위 이상은 ,가 있어서 오류 -> replace써서 ,없앰

for row in data_2021:
  if '한솔동' in row[0]:
    for i in row[3:]:
      result_2021.append(int(i.replace(',','')))


plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.title('세종 한솔동 인구구조')
plt.style.use('ggplot')
plt.plot(result_2012,color='k',label='2012')
plt.plot(result_2021,color='r',label='2021')
plt.legend()
plt.show()
```
