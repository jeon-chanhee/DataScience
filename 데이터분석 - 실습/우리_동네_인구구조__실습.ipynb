```python
import csv
import matplotlib.pyplot as plt



f = open('age.csv',encoding='cp949')
data = csv.reader(f)

result = []

for row in data:
  if '산남동' in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))  # -> 1000단위 이상은 ,가 있어서 오류 -> replace써서 ,없앰

print(result)
plt.figure(dpi=200)
plt.title('청주 산남동 인구구조')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```
