[Step 1] 데이터 준비 - read_csv() 함수로 자동차 연비 데이터셋 가져오기

```python
# csv 파일을 데이터프레임으로 변환
df = pd.read_csv('auto-mpg.csv',header=None)
df
```

```python
# 열 이름 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

데이터 살펴보기

```python
df.head()
```

[Step 2] 데이터 탐색 및 데이터 전처리

```python
# 데이터 자료형 확인
df.info()
```

```python
# horsepower 열의 자료형 (문자열 -> 숫자)
df['horsepower'].unique()
```

```python
# '?'을 np.nan으로
df['horsepower'].replace('?',np.nan,inplace=True)
```

```python
# 누락 데이터를 삭제
df.dropna(subset=['horsepower'], axis=0, inplace=True)
```

```python
df.info()
```

```python
# 문자열 -> 숫자로 변환
df['horsepower'] = df['horsepower'].astype('float')
```

```python
df.info()
```

[Step 3] 속성(feature 또는 variable) 선택

```python
# 분석에 활용할 속성을 선택 (연비, 실린더, 마력, 중량)
ndf = df[['mpg','cylinders','horsepower','weight']]
ndf.head()
```

[Step 4] 데이터셋 구분 - 훈련용(train data)/검증용(test data)

```python
# 속성(변수) 선택
X = ndf[['cylinders','horsepower','weight']]  # 독립변수는 2차원으로 만들어야 해서 [] 2개 써야함 왜? 특성 여러가지이기 때문
y = ndf['mpg'] # 종속 변수 y
```

```python
# train data 와 test data로 구분 (7:3 비율)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,                   # 독립 변수
                                                    y,                   # 종속 변수
                                                    test_size=0.3,       # 검증 30% 
                                                    random_state=7)     # 랜덤 추출 값
```

[Step 5] Ridge 회귀분석 모형 - sklearn 사용

```python
# sklearn 라이브러리에서 선형회귀분석 모듈 가져오기
from sklearn.linear_model import Ridge

# Ridge 회귀분석 모형 객체 생성 후 학습 시키기
rdg = Ridge(alpha=1).fit(X_train, y_train)
```

[Step 6] 성능 평가하기

```python
# 학습을 마친 모형에 test_data를 적용하여 결정계수(R-제곱) 계산 -> 
# 결정계수 값이 클수록 모형의 예측 능력이 좋다고 판단
r_square = rdg.score(X_test,y_test)
r_square

# alpha = 500
#훈련 세트 정확도 : 70.54%
#테스트 세트 정확도 : 69.79%

# alphat = 10
# 훈련 세트 정확도 : 70.71%
# 테스트 세트 정확도 : 69.45%

# alphat = 0.01
# 훈련 세트 정확도 : 70.71%
# 테스트 세트 정확도 : 69.39%

# alpha = 30
# 훈련 세트 정확도 : 70.70%
# 테스트 세트 정확도 : 69.54%
```

```python
# 모형에 전체 X 데이터를 입력하여 예측한 값 y_hat으로 담기
y_hat = rdg.predict(X)
```

```python
# 모형의 테스트셋 X_test 데이터를 입력하여 예측한 값 y_hat_test로 담기
y_hat_test = rdg.predict(X_test)
```

```python
# alpha 값에 따른 정확도 저장할 리스트 변수
train_accuracy = []
test_accuracy = []

# 1에서 300까지 n_alpha를 적용
alpha_settings = range(1,300)

for n_alpha in alpha_settings:
  
  # 모델 생성
  rdg = Ridge(alpha=n_alpha).fit(X_train, y_train)

  # 훈련 세트 정확도 저장
  train_accuracy.append(rdg.score(X_train,y_train))

  # 테스트 세트 정확도 저장
  test_accuracy.append(rdg.score(X_test,y_test))

# 훈련 vs 테스트 정확도 시각화
plt.figure(dpi=100)
plt.rc('font', family='NanumBarunGothic')

plt.plot(alpha_settings, train_accuracy, label='훈련 정확도')
plt.plot(alpha_settings, test_accuracy, label='테스트 정확도')
plt.ylabel('정확도')
plt.xlabel("n_alpha")
plt.legend()
plt.show()
```

```python
# train 데이터(X_train) 의 산점도와 test 데이터로 예측한(y_hat_test) 회귀선을 그래프로 출력

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
ax.plot(X_train, y_train, 'o', label='Train Data') # 데이터의 분포
ax.plot(X_test, y_hat_test,'r+', label='Predicted Value') # 모형이 예측한 회귀선
ax.legend(loc='best') # 최적의 공간에 범례 배치함
plt.xlabel('features')
plt.ylabel('mpg')
plt.show()
```

```python
print('훈련 세트 정확도 : {:.2f}%'.format(rdg.score(X_train,y_train)*100))
print('테스트 세트 정확도 : {:.2f}%'.format(rdg.score(X_test,y_test)*100))
```
