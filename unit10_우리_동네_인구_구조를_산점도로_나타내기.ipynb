꺾은선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('찾고 싶은 지역의 이름을 알려주세요')

for row in data:
  if name in row[0]:
    for i in range(0,101):
      m.append(int(row[i+3].replace(',','')))
      f.append(int(row[-(i+1)].replace(',','')))
    break   # 뒤에 있는 같은 이름의 데이터가 있더라도 1개만 추가하고 for문 탈출할거야!

f.reverse()

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')

# 꺾은선 그래프로 남여 데이터 표현하기
plt.plot(m, label='남성')
plt.plot(f, label='여성')
plt.legend()
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

result =[] # 남성 인구 - 여성 인구 값 담을 리스트 변수 선언

name = input('찾고 싶은 지역의 이름을 알려주세요')
for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '지역의 남녀 성별 인구 차이 분포')
plt.bar(range(101), result)
plt.show() 
```

제주도의 연령대별 성별 비율을 산점도로 표현하기

남녀 연령별 인구 데이터 scatter() 함수로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('찾고 싶은 지역의 이름을 알려주세요')

for row in data:
  if name in row[0]:
    for i in range(0,101):
      m.append(int(row[i+3].replace(',','')))
      f.append(int(row[-(i+1)].replace(',','')))
    break   # 뒤에 있는 같은 이름의 데이터가 있더라도 1개만 추가하고 for문 탈출할거야!

f.reverse()

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')

# 산점도 그래프로 남여  인구수 데이터 표현하기
plt.scatter(m,f) # x 축은 남자 y 축은 여자
plt.show()
```

원의 크기는 인구수의 제곱근 추가

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

m = []
f = []
size = [] # 인구수 사이즈 리스트 변수 선언

name = input('찾고 싶은 지역의 이름을 알려주세요')

for row in data:
  if name in row[0]:
    for i in range(0,101):
      m.append(int(row[i+3].replace(',','')))
      size.append(math.sqrt(int(row[i+3].replace(',',''))))  # sqrt 제곱근  -> 값이 큰값을 작은 사이즈로 바꿔서 비교쉽게
      f.append(int(row[-(i+1)].replace(',','')))
    break   # 뒤에 있는 같은 이름의 데이터가 있더라도 1개만 추가하고 for문 탈출할거야!

f.reverse()

# 인구수와 sqrt(인구수 차이확인)
print(m)
print(size)

# 산점도 그래프로 남여  인구수 데이터 표현하기
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.scatter(m,f,s=size,c=range(101),cmap='jet',alpha=0.5) # x 축은 남자 y 축은 여자
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g') # 추세선 그리기
plt.show()
```
