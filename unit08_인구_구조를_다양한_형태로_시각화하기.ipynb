지역명을 입력받아서 연령별 인구수 데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv',encoding='cp949')
data = csv.reader(f)

result = []   # 0~100세 이기에 순서대로 출력하고 싶어서

# 지역명을 입력받는 변수 선언
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))


plt.figure(dpi=200)
plt.style.use('ggplot')
plt.title(name +'지역의 인구 구조')
plt.plot(result)
plt.show()
```

성별 데이터 저장하기(Type 1)

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in range(0,101):
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)]))     # for 문을 한번에 돌리고 싶어서 이와 같이 코딩 0~101의 range가 있는데 남성의 경우 3~103까지가 인덱스
                                     # 여자의 경우 -101 ~ -1 까지가 범위임

f.reverse() # 0세 부터 인구수 나올 수 있도록 데이터 뒤집기

print(m)
print(f)
```

성별 데이터 순서대로 저장하기(Type2)

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]: # 남성 데이터 : 3 ~ 103번 슬라이싱 - 정수데이터로 변경 - m 리스트에 저장
      m.append(int(i))
    for i in row[106:]:  #여성 데이터 : 106 ~ 끝까지 데이터 슬라이싱 - 정수 데이터로 변경 - f 리스트에 저장
      f.append(int(i))

print(m)
print(f)
```

데이터 시각화하기

```python
plt.figure(dpi=200)
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

항아리 그래프를 위해 남성 데이터 음수로 바꾸기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]: # 남성 데이터 : 3 ~ 103번 슬라이싱 - 정수데이터로 변경 - m 리스트에 저장
      m.append(int(i))  # - 붙이기 
    for i in row[106:]:  #여성 데이터 : 106 ~ 끝까지 데이터 슬라이싱 - 정수 데이터로 변경 - f 리스트에 저장
      f.append(-int(i))

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('신도림 지역의 남녀 성별 인구 분포도')
plt.barh(range(101), m, color = 'skyblue',label = '남성')
plt.barh(range(101), f, color = 'hotpink', label = '여성')
plt.legend()
plt.show()
```
